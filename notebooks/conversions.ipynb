```python
import pandas as pd

# Read txt
df = pd.read_csv("../data/test_FD001.txt", sep=r"\s+", header=None)

# Remove the last two empty columns
df = df.iloc[:, :26]

# Column names
columns = [
    "engine_id",
    "cycle",
    "op_setting_1",
    "op_setting_2",
    "op_setting_3",
]

columns += [f"sensor_{i}" for i in range(1, 22)]

df.columns = columns

# Save CSV
df.to_csv(
    "test_FD001.csv",
    index=False
)

print(df.head())
```
